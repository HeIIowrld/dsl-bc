사내 LLM 회귀테스트용 지식 베이스를 만들기 위한 크롤링 실행 노트북입니다.

- 각 섹션은 출처별로 분리되어 있습니다.
- 기본값은 긴 크롤링이 자동 실행되지 않도록 `RUN_CRAWL = False`입니다.
- 전체 재수집이 필요하면 설정 셀에서 `RUN_CRAWL = True`로 바꾸고 필요한 섹션만 실행하세요.
- 마지막 통합 셀은 `llm_regression_all_sources.csv/jsonl`을 다시 생성합니다.


## 0. 공통 설정

```python
from pathlib import Path
import csv
import json
import subprocess
import sys
from collections import Counter

cwd = Path.cwd()
if (cwd / "sources" / "bc_cs_notice").exists():
    PROJECT_ROOT = cwd
    CRAWL_DIR = cwd / "sources" / "bc_cs_notice"
elif cwd.name == "bc_cs_notice":
    CRAWL_DIR = cwd
    PROJECT_ROOT = cwd.parent.parent if cwd.parent.name == "sources" else cwd.parent
else:
    raise RuntimeError(f"작업 위치를 확인할 수 없습니다: {cwd}")

PYTHON = PROJECT_ROOT / ".venv" / "Scripts" / "python.exe"
if not PYTHON.exists():
    PYTHON = Path(sys.executable)

SCRIPTS_DIR = CRAWL_DIR / "scripts"
OUT_DIR = CRAWL_DIR / "out"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# 긴 네트워크 크롤링 실행 스위치입니다.
RUN_CRAWL = False

# BC카드 저장 HTML seed를 다시 쓰려면 파일 경로를 지정하세요.
# 예: BC_SEED_HTML = CRAWL_DIR / "보험모집종사자 조회_고객센터.html"
BC_SEED_HTML = None

def run_script(script_name: str, *args: object) -> subprocess.CompletedProcess:
    cmd = [str(PYTHON), str(SCRIPTS_DIR / script_name), *map(str, args)]
    print("$", " ".join(cmd))
    result = subprocess.run(cmd, cwd=PROJECT_ROOT, text=True, capture_output=True)
    if result.stdout:
        print(result.stdout)
    if result.stderr:
        print(result.stderr)
    result.check_returncode()
    return result

def jsonl_count(path: Path) -> int:
    if not path.exists():
        return 0
    with path.open(encoding="utf-8") as f:
        return sum(1 for line in f if line.strip())

print("PROJECT_ROOT:", PROJECT_ROOT)
print("CRAWL_DIR:", CRAWL_DIR)
print("SCRIPTS_DIR:", SCRIPTS_DIR)
print("OUT_DIR:", OUT_DIR)
print("PYTHON:", PYTHON)
```

## 10. 산출물 검증

통합 corpus의 row 수, 출처별 분포, 빈 content 여부를 확인합니다.

```python
combined_jsonl = OUT_DIR / "llm_regression_all_sources.jsonl"
combined_csv = OUT_DIR / "llm_regression_all_sources.csv"

rows = []
with combined_jsonl.open(encoding="utf-8") as f:
    for line in f:
        if line.strip():
            rows.append(json.loads(line))

print("combined_jsonl:", combined_jsonl)
print("rows:", len(rows))
print("blank content:", sum(1 for row in rows if not str(row.get("content", "")).strip()))
print("source_type:")
for source_type, count in Counter(row.get("source_type", "") for row in rows).most_common():
    print(f"  {source_type}: {count}")

print("\nfiles:")
for path in sorted(OUT_DIR.glob("*_corpus.jsonl")):
    print(f"  {path.name}: {jsonl_count(path)}")
```

## 11. 빠른 샘플 확인

출처별로 첫 문서의 제목과 본문 앞부분을 확인합니다.

```python
seen = set()
for row in rows:
    source_type = row.get("source_type", "")
    if source_type in seen:
        continue
    seen.add(source_type)
    print("=" * 80)
    print("source_type:", source_type)
    print("title:", row.get("title"))
    print("url:", row.get("url"))
    print(str(row.get("content", ""))[:500])
```
